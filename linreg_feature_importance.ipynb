```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.metrics import *
import matplotlib.pyplot as plt
```

```python
def gen_data(noise=False):
    df = pd.DataFrame({
        'x1': np.random.randint(1, 10, 100), 
        'x2': np.random.randint(1, 10, 100),
        'x3': np.random.randint(1, 10, 100),
        'x4': np.random.randint(1, 10, 100),
    })

    df['y'] = df.x1 + df.x2
    
    if noise:
        df["y"] += np.random.normal(0, 5, 100)
    
    return df
```

```python
df = gen_data()

df.head()
```

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
from sklearn.linear_model import LinearRegression

features = ["x1", "x2", "x3", "x4"]

for f in features:
    print("feature: ", f)
    
    X = df[f].values.reshape(-1,1)

    model = LinearRegression()
    model.fit(X, df.y)
    model.score(X, df.y)
    y_pred = model.predict(X)

    print("model: ", model.coef_, model.intercept_)
    print("R_squared: ", r2_score(df.y, y_pred))
    print("MAE: ", mean_absolute_error(df.y, y_pred))
    print("MSE: ", mean_squared_error(df.y, y_pred))
    
    plt.figure()
    
    plt.plot(X, df.y, "ob")
    plt.plot(X, model.coef_ * X + model.intercept_, "-r")
```

```python
from sklearn.linear_model import LinearRegression
from itertools import permutations

features = ["x1", "x2", "x3", "x4"]

for feature in permutations(features, 2):
    feature = list(feature)
    print("feature: ", feature)
    
    X = df[feature].values

    model = LinearRegression()
    model.fit(X, df.y)
    model.score(X, df.y)
    y_pred = model.predict(X)

    print("model: ", model.coef_, model.intercept_)
    print("R_squared: ", r2_score(df.y, y_pred))
    print("MAE: ", mean_absolute_error(df.y, y_pred))
    print("MSE: ", mean_squared_error(df.y, y_pred))

    plt.figure()

    plt.plot(X[:,0], df.y, 'ob')
    plt.plot(X[:,0], model.predict(X), 'or')
    
    plt.title(f"Feature: {feature}, R2: {model.score(X, df.y)}, coef: {model.coef_}")
```
